```python
import json
import math
import numpy as np
```

```python
gene2cpg = './gene2cpg.json'
cpg2gene = './cpg2gene.json'
cpgname = './cpgname.txt'
samplesfile = './samples.txt'
```

```python
with open(gene2cpg,'r') as f:
    g2c_dict=json.load(f)
with open(cpg2gene,'r') as f:
    c2g_dict=json.load(f)
with open(cpgname,'r') as f:
    names = f.read().splitlines()
with open(samplesfile,'r') as f:
    onesample = f.readline().split()
```

```python
label = onesample.pop(-1)
d = {}
for ind, feature in enumerate(onesample):
    if names[ind] in c2g_dict:
        for gene in c2g_dict[names[ind]]:
            if gene not in d:
                d[gene]=[]
            d[gene].append((names[ind],float(feature)))
```

```python
def analyze_dict(d):
    stavar_list = []
    m = len(d)
    NA_rate = [0]*21
    distri_rate = [0]*11
    for v in d.values():
        vlist = []
        l = len(v)
        count = 0
        for pair in v:
            if pair[1] == -1.0:
                count += 1
            else:
                vlist.append(pair[1])
        varray = np.array(vlist)
        me = np.mean(varray)
        sta_var = math.sqrt(np.var(varray))
        stavar_list.append(sta_var)
        rate = math.ceil(count*20.0/l)
        qnum = np.sum((varray>me-sta_var) & (varray<me+sta_var))
        qrate = math.floor(qnum*10.0/l)
        NA_rate[rate] += 1
        distri_rate[qrate] += 1
    print('missing rate distribution: first no missing, 5 percent each interval left excluded right included')
    print(NA_rate) 
    print('in one sigma distribution: last all, 5 percent each interval left excluded right included')
    print(distri_rate) 
    return stavar_list
```

```python
a=analyze_dict(d)
```
